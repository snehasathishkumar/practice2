```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
import google.generativeai as genai
from langchain_community.vectorstores import FAISS
from dotenv import get_key
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
text = """ 
Paris is the best. 

Sneha Sathish Kumar. Recursive character text splitter sneha sathish kumar.

Sneha sathish Kumar Sneha Sathish Kumar. Recursive character text splitter sneha sathish kumar. Sneha Sathish Kumar. Recursive character text splitter sneha sathish kumar.
"""
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 0,
    length_function =len, 
    separators=["\n",".",""]
)
```

```python
chunks = text_splitter.create_documents(text)
chunks
```

```python
len(chunks[2])
```

```python
# embeddings = genai.generate_embeddings("models/embedding-004")
api_key = get_key(".env", key_to_get="GEMINI_API_KEY")
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-004",google_api_key=api_key)
```

```python
db = FAISS.from_documents(chunks, embeddings)
```
